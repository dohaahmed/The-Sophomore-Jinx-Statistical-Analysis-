```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_rel
```

```python
file_path = "Hitter Sheet Example.csv"
data = pd.read_csv(file_path)
```

```python
df = data.dropna(subset=['Sophomore BA', 'Sophomore SLG']).copy()
df['BA_Diff'] = df['Sophomore BA'] - df['Rookie BA']
df['SLG_Diff'] = df['Sophomore SLG'] - df['Rookie SLG']
```

```python
t_ba, p_ba = ttest_rel(df['Sophomore BA'], df['Rookie BA'])
t_slg, p_slg = ttest_rel(df['Sophomore SLG'], df['Rookie SLG'])
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['BA_Diff'], bins=15, color='skyblue', edgecolor='black')
plt.axvline(0, color='red', linestyle='--')
plt.title("Distribution of Change in Batting Average (Sophomore - Rookie)", fontsize=13)
plt.xlabel("Change in BA")
plt.ylabel("Number of Players")
plt.grid(alpha=0.3)
plt.show()

plt.figure(figsize=(8,5))
plt.hist(df['SLG_Diff'], bins=15, color='lightgreen', edgecolor='black')
plt.axvline(0, color='red', linestyle='--')
plt.title("Distribution of Change in Slugging Average (Sophomore - Rookie)", fontsize=13)
plt.xlabel("Change in SLG")
plt.ylabel("Number of Players")
plt.grid(alpha=0.3)
plt.show()
```
